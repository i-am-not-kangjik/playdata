```python
# 아래 예제는 label 키워드 인수를 통해 legend label 설정하고
# legend(loc) 인자에 2를 전달해 좌측 상단에 위치하게 명시하고 있다.

X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2) # upper left
plt.show()
```

```python
# 위/아래 두 개의 플롯을 하나의 Figure 안에 그리려면 다음처럼 명령을 실행해야 한다.
# 여기서 숫자 인덱싱은 파이썬이 아닌 Matlab 관행을 따르기 때문에 첫번째 플롯을 가리키는 숫자가 0이 아니라 1임에 주의해야 한다.
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.title('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
# 만약 2x2 형태의 네개의 플롯이라면 다음과 같이 그린다.
# subplot의 인수는 (2, 2, 1)을 줄여서 221라는 하나의 숫자로 표시할 수 있다.

np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# subplots()에 인수를 두개만 전달하면 복수의 Axes 객체를 동시에 생성할 수도 있다.
# 이 때는 2차원 ndarray 형태로 Axes 객체가 반환된다.
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

### 연습 문제
타이타닉호 승객을 '미성년자', '청년', '장년', '중년', 노년' 나이 그룹으로 나눈다.   
그리고 각 나이 그룹으로 승객 비율을 파이 차트로 출력해보라.

```python
import pandas as pd
titanic = sns.load_dataset("titanic")
titanic.age
```

```python
titanic.dropna(subset=["age"], inplace=True)
```

```python
titanic.age.count()
```

```python
bins = [0, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "장년", "중년", "노년"]
cats = pd.cut(titanic.age, bins, labels=labels)
cats.value_counts()
```

```python
explode = (0.1, 0, 0, 0, 0)
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral", "orange"]
labels = ["장년", "청년", "미성년자", "중년", "노년"]
plt.title("승객 연령")
plt.pie(cats.value_counts(), explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis = 'equal'
plt.show()
```

```python
def f(x, y):
    return(1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Countour plots")
plt.contourf(XX, YY, ZZ, alpha=0.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title('3D surface Plot')
ax.plot_surface(XX, YY,ZZ, rstride = 1, cstride = 1, cmap = 'hot')
plt.show()
```
